### Ε1: Εισόδημα και Αποτελέσματα των Τεστ

Θα διερευνήσετε τη σχέση μεταξύ του εισοδήματος σε κάθε χώρα και των αποτελεσμάτων των τεστ TIMSS και PIRLS, όπως στο 4.1 και την εικόνα 5 της μελέτης. Συγκεκριμένα, θα διερευνήσετε μέσω παλινδρόμησης τη σχέση μεταξύ του λογαρίθμου του εισοδήματος και του σκορ TIMMS, και του λογαρίθμου του εισοδήματος και του σκορ PIRLS. Αφού φτιάξετε  τα σχετικά μοντέλα και σχολιάσετε τα αποτελέσματά σας, θα πρέπει να παράξετε διαγράμματα των δεδομένων αντίστοιχα με τα παρακάτω.

Για την απάντηση στην ερώτηση θα χρειαστείτε και το αρχείο `WDI_data.csv`, το οποίο περιέχει δεδομένα εισοδήματος ανά χώρα από την Παγκόσμια Τράπεζα.

![GDP (PPP) and TIMSS](gdp_timss.svg)

![GDP (PPP) and PIRLS](gdp_pirls.svg)

```python
import pandas as pd

country_scores = pd.read_excel("patel-sandefur-human-capital-final-results.xlsx",sheet_name=1)

country_scores = country_scores.dropna(axis=0)

country_scores
```

```python
wdi = pd.read_csv("WDI_data.csv")

wdi = wdi.dropna()

wdi = wdi[wdi["2015 [YR2015]"]!=".."]

wdi = wdi[wdi["2015 [YR2015]"].astype(float)>1000]

wdi
```

```python
import numpy as np

newdf = country_scores.merge(wdi, on="Country Code", how="inner")

PIRLS = newdf["Median Reading Score (PIRLS Scale, 4th Grade)"].values

TIMSS = newdf["Median Math Score (TIMSS Scale, 4th Grade)"].values

y = newdf["2015 [YR2015]"]

y = y.astype(float).values

y = np.log(y)
```

```python
import seaborn as sns

sns.scatterplot(x=y,y=PIRLS)
```

```python
sns.scatterplot(x=y,y=TIMSS)
```

```python
import sklearn as sk
from sklearn import linear_model

from matplotlib import pyplot as plt

lr = sk.linear_model.LinearRegression()

y = y.reshape(-1,1)

lr.fit(y,PIRLS)

y_pred = lr.predict(y)

plt.scatter(y,PIRLS)
plt.plot(y,y_pred,color="r")
plt.show()
```

```python
lr.fit(y,TIMSS)

y_pred = lr.predict(y)

plt.scatter(y,TIMSS)
plt.plot(y,y_pred,color="r")
plt.show()
```
